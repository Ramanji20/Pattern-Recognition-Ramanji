```python
# Check the shape of the identity_matrix and img_average_channel
identity_shape = identity_matrix.shape
img_shape = img_average_channel.shape
T = identity_matrix[rows]

print(T.shape)
print("\nT is a matrix that represents a specific way to shuffle or permute the rows of another matrix. Each row in T corresponds to a row index, and its position in T determines the new order in which the rows should be arranged when multiplying it with another matrix.")
if T.shape != identity_matrix.shape or T.shape != img_average_channel.shape:
    T = T.reshape(identity_matrix.shape)
    print("\nReshaping T")
else:
    print("\nT has the same shape")
```

## Part 6

```python
# Assuming T and img_average_channel are defined as per previous steps
transformed_image = np.dot(T, img_average_channel)

# Check the shape of the resulting transformed_image
print("Shape of Transformed Image:", transformed_image.shape)

# Plot the transformed_image
plt.imshow(transformed_image)
plt.title("Transformed Image")
plt.axis('off')
plt.show()
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
# Check if T is invertible by computing its determinant
det_T = np.linalg.det(T)

# Check if the determinant is non-zero
if det_T != 0:
    # Calculate the inverse of T
    Tinv = np.linalg.inv(T)
    
    # Perform the inverse operation to recover the original image
    original_image = np.dot(Tinv, transformed_image)
    
    # Plot the original image
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
else:
    print("Matrix T is not invertible (singular), so the original image cannot be recovered.")
```

## Part 8: Extra Points

* We want to perform the same operations but on the columns of the original image. How would you do the matrix algebra? Plot the scrambled image for the proof of concept.


```python
# Get the number of columns in img_average_channel
M = img_average_channel.shape[1]

# Create an array containing numbers from 0 to M-1 and shuffle it for columns
columns = np.arange(M)
np.random.shuffle(columns)

# Create a permutation matrix for columns
C = np.eye(M)[:, columns]

# Multiply the original image by the permutation matrix for columns
scrambled_image = np.dot(img_average_channel, C)

# Check the shape of the scrambled image
print("Shape of Scrambled Image:", scrambled_image.shape)

# Plot the scrambled image
plt.imshow(scrambled_image)
plt.title("Scrambled Image")
plt.axis('off')
plt.show()
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Get the number of rows and columns in img_average_channel
N, M = img_average_channel.shape

# Create permutation arrays for rows and columns and shuffle them
rows = np.arange(N)
np.random.shuffle(rows)

columns = np.arange(M)
np.random.shuffle(columns)

# Create permutation matrices for rows and columns
R = np.eye(N)[rows, :]
C = np.eye(M)[:, columns]

# Scramble both rows and columns of the original image
scrambled_image = np.dot(np.dot(R, img_average_channel), C)

# Check the shape of the scrambled image
print("Shape of Scrambled Image:", scrambled_image.shape)

# Plot the scrambled image
plt.imshow(scrambled_image)
plt.title("Scrambled Image")
plt.axis('off')
plt.show()

# Invert the scrambling operations to recover the original image
Cinv = np.linalg.inv(C)
Rinv = np.linalg.inv(R)

# Recover the original image
recovered_image = np.dot(np.dot(Rinv, scrambled_image), Cinv)

# Plot the recovered original image
plt.imshow(recovered_image)
plt.title("Recovered Original Image")
plt.axis('off')
plt.show()
```
